# Bayesian paradigm 

We are going to run a setting in which we have many observations in order to estimate the click through rate following the bayesian approach.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
def plot(a, b, trial, ctr):
  x = np.linspace(0, 1, 200)
  y = beta.pdf(x, a, b)
  mean = float(a) / (a + b)
  plt.plot(x, y)
  plt.title("Distribution after %s impressions, true rate = %.1f, mean = %.2f" % (trial, ctr, mean))
  plt.show()
```

```python
true_ctr = 0.3
a, b = 1, 1 # beta parameters
show = [0, 5, 10, 25, 50, 100, 200, 300, 500, 700, 1000, 1500]
for t in range(1501):
  click_result = (np.random.random() < true_ctr)
  if click_result:
    a += 1
  else:
    b += 1

  if t in show:
    plot(a, b, t+1, true_ctr)
```

As we increased the number of impressions -i.e. we collect more data-, the variance of the distribution decreases and the shape becomes skinnies because we are now more certain of the value of the parameter. In Bayesian statistics, the confidence interval is sort of built into the posterior distribution.
